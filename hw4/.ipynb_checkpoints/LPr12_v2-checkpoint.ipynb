# The task

Goal: prediction of the Prognosis column
Evaluation metric: accuracy

DATA DESCRIPTION
Predict the severity of COVID-19 in patients of Italian hospitals. The data is collected from 6 hospitals. Chest X-ray images are provided for each patient, which are collected in imgs.npy. 

**Features**

* Age - Patient’s age (years)
* Sex - Patient’s sex (0 – male, 1 – female)
* Body Temperature (°C) - Patient temperature at admission (in °C)
* Cough - Cough
* Dyspnea - Patient had intense tightening in the chest, air hunger, difficulty breathing, breathlessness or feeling of suffocation
* WBC - White blood cells count (10^9/L)
* CRP - C-reactive protein concentration (mg/dL)
* Fibrinogen - Fibrinogen concentration in blood (mg/dL)
* LDH - Lactate dehydrogenase concentration in blood (U/L)
* D-dimer - D-dimer amount in blood
* O2 - Oxygen percentage in blood
* PaO2 - Partial pressure of of oxygen in arterial blood (mmHg)
* SaO2 - Arterial oxygen saturation (%)
* pH - Blood pH
* Cardiovascular Disease - Patient had cardiovascular disease
* Respiratory Failure - Patient had respiratory failure
* Image - Index of corresponding image in imgs.npy (starting from 0)
* Death - Whether the patient died (not provided in test column)
* Prognosis - Severity of COVID

## Hyperparameter tuning

```python
C_range_1 = np.linspace(0.1,0.9,9)
C_range_2 = np.linspace(1.0,9.0,9)
C_range_3 = np.linspace(10.0,90,9)
C_range_4 = np.linspace(100.0,900,9)
```

```python
C_range = np.hstack([C_range_1,C_range_2,C_range_3,C_range_4])
```

```python
C_range.shape
```

```python
C_range
```

```python
# simple grid search
train_accs = []
valid_accs = []

for C in C_range:
    clf = SVC(C=C,kernel="linear").fit(X_train,y_train)
    y_pred_train = clf.predict(X_train)
    y_pred_test = clf.predict(X_test)
    
    train_accs.append(accuracy_score(y_train,y_pred_train))
    valid_accs.append(accuracy_score(y_test,y_pred_test))
    
    print(f"C: {C:.2f} | TA {train_accs[-1]:.4f} VA {valid_accs[-1]:.4f}")
    
```

```python
plt.figure(figsize=(7,7))
plt.plot(train_accs,label="Training accuracies")
plt.plot(valid_accs,label="Validation accuracies")
plt.legend()
plt.show()
```

# Deep learning

```python
import torch
import torchvision
import torch.nn as nn
from torch import optim
import torch.nn.functional as F
from time import time
```

```python
import torchmetrics
```

```python
class CustomDataset(torch.utils.data.Dataset):
    def __init__(self, tabular, target, img_idcs, transforms = None, img_channel_number = 3):
        self.tabular = tabular
        self.target = target
        self.img_idcs = img_idcs
        self.transforms = transforms
        self.num_channels = img_channel_number

    def __len__(self):
        return len(self.target)

    def __getitem__(self, idx):
        tabular = torch.FloatTensor(self.tabular.iloc[idx])
        target = torch.FloatTensor([self.target.iloc[idx]])
        
        img_idx = self.img_idcs.iloc[idx] # get the index of the picture (the idx-th patient)
        img = torch.FloatTensor(img_array[img_idx])
        img = torch.stack([img for _ in range(self.num_channels)]) # grey-scaled
        
        if self.transforms:
            img = self.transforms(img)
            
        return {"img":img,"target":target,"tabular":tabular}
```

```python
train_transforms = torchvision.transforms.Compose([
    torchvision.transforms.Normalize([0.5, 0.5, 0.5],[0.25,0.25,0.25]),
    torchvision.transforms.RandomRotation(degrees=15)
])
valid_transforms = train_transforms = torchvision.transforms.Compose([
    torchvision.transforms.Normalize([0.5, 0.5, 0.5],[0.25,0.25,0.25])
])
train_dataset = CustomDataset(X_train,y_train,image_indices_train,transforms=train_transforms)
valid_dataset = CustomDataset(X_test,y_test,image_indices_test,transforms=valid_transforms)

# create batches from them
train_data = torch.utils.data.DataLoader(train_dataset,shuffle=True,batch_size=32)
valid_data = torch.utils.data.DataLoader(valid_dataset,shuffle=False,batch_size=32)
```
